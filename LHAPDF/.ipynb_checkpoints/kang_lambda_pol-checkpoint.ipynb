```python
##### Grab a FF

zD_Lambda = lhapdf.getPDFSet('DSS-1997-Lambda')
zD_Lambda = zD_Lambda.mkPDFs()
zD_Lambda[0].xfxQ(2,0.1,10.)
```

```python
### Look at Info File

# Error Type: replicas
# 0th entry everage? Not Specified
# quark index up to 5, not 6 >> No bottom quark
# ^ But doesn't throw an error
```

```python
### Collect Lambda FFs for fixed Q

zvalues = np.logspace(-3,0,100)
qvalues = np.logspace(0,2,500)

### Q = 10 GeV

# xf0[0].xfxQ(0,0.1,10)

Q0 = 91.2
zD_Lambda_FFs = []
Nz = len(zvalues)

quark_index = [1,2,3,4,5,6,-1,-2,-3,-4,-5,-6,21]

for index in quark_index:

    FF = np.zeros(Nz) # Sets last value to 0
    
    for i in range(Nz-1):
    
        z = zvalues[i]
    
        FF[i] = zD_Lambda[0].xfxQ(index,z,Q0)

    # # For gluon
    # if index == 21:

    #     FF = FF / 10

    zD_Lambda_FFs.append(FF)
```

```python
### Check bottom quark index

zD_Lambda_FFs[-6]
```

```python
### Plot all PDFS for fixed Q = 10 GeV
### Note pdfs return zD not D

plt.figure(0)

# for i in range(len(pdfs)):
for i in range(len(quark_index)):

    plt.plot(zvalues, zD_Lambda_FFs[i]/zvalues, label = str(quark_index[i]))

# plt.xscale('log')
plt.yscale('log')
# plt.ylim((-0.01,2.0))
plt.xlim((zvalues[0], zvalues[-1]))

plt.ylabel(r'$zD_{\Lambda}(z)$')
plt.xlabel('$z $',fontsize=18)
plt.grid()
plt.legend()
```

```python
### Check with Figure 30 of https://arxiv.org/pdf/0803.2768

# Looks Good!

zD_Lambda_uubar =  zD_Lambda_FFs[1] + zD_Lambda_FFs[7]


plt.figure()
plt.plot(zvalues, zD_Lambda_uubar / zvalues, label = r'$u + \overline{u}$')

# plt.xscale('log')
plt.yscale('log')
# plt.ylim((-0.01,2.0))
plt.xlim((zvalues[0], zvalues[-1]))

plt.ylabel(r'$zD_{\Lambda}(z)$')
plt.xlabel('$z $',fontsize=18)
plt.grid()
plt.legend()
```
