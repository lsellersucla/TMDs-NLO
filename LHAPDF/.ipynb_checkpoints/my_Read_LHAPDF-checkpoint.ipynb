```python
##### Plot quarks from first set

xvalues = np.logspace(-3,0,100)
qvalues = np.logspace(0,2,500)

### Q = 10 GeV

# xf0[0].xfxQ(0,0.1,10)

# Q0 = 100.0
Q0 = 10.0
pdfs = []
Nx = len(xvalues)

quark_index = [1,2,3,4,5,6,-1,-2,-3,-4,-5,-6,21]

for index in quark_index:

    pdf = np.zeros(Nx) # Sets last value to 0
    
    for i in range(Nx-1): # No -1 throws an error >> Must not like x = 1
    
        x = xvalues[i]
    
        pdf[i] = xf0[0].xfxQ(index,x,Q0)

    # # For gluon
    # if index == 21:

    #     pdf = pdf / 10

    pdfs.append(pdf)
```

```python
##### FFs for Fixed Q

zvalues = np.logspace(-2,0,100)
qvalues = np.logspace(0,2,500)

### Q = 5 GeV

# xf0[0].xfxQ(0,0.1,10)

Q0 = 10.0
FFs = []
Nz = len(zvalues)

quark_index = [1,2,3,4,5,6,-1,-2,-3,-4,-5,-6,21]

for index in quark_index:

    FF = np.zeros(Nz) # Sets last value to 0
    
    for i in range(Nz): # No -1 throws an error >> Must not like x = 1
    
        z = zvalues[i]
    
        FF[i] = zPIsum[0].xfxQ(index,z,Q0)

    # # For gluon
    # if index == 21:

    #     pdf = pdf / 10

    FFs.append(FF)
```

```python
### Locate Indices

iq = np.where(np.array(quark_index) == 2)[0][0]
iqbar = np.where(np.array(quark_index) == -2)[0][0]
print(iq, iqbar)
```

```python
### Plot FF 
#
## I think this is u-quark to pion

## D^{\pi^{\pm}}_{u^{+}}(z) is sum of pi + and pi - and sum of q and qbar

plt.plot(zvalues, FFs[iq]+FFs[iqbar], label = str(quark_index[1]))

plt.xscale('log')
# plt.ylim((-0.01,2.0))
plt.xlim((zvalues[0], zvalues[-1]))

plt.ylabel(r'$zD^{\pi^{\pm}}_{u^{+}}(z)$', fontsize = 18)
plt.xlabel('$z $',fontsize=18)
plt.grid()
plt.title('NNLO', fontsize = 20)
plt.legend()
```

```python
### Try to find how many Replicas (Should be in info file?)

# Info File >> numMembers = 101

# max_i = 101 # Looks like 101 is max
# for i in range(max_i):

#     print(zPIsum[i].xfxQ(2,0.1,5.), i)
```

```python
### Compile all Replicas

from tqdm import tqdm

numReps = 101

reps = []

for n in tqdm(range(numReps)):

    FFs = []

    for index in quark_index:
    
        FF = np.zeros(Nz) # Sets last value to 0
        
        for i in range(Nz): # No -1 throws an error >> Must not like x = 1
        
            z = zvalues[i]
        
            FF[i] = zPIsum[n].xfxQ(index,z,Q0)
    
        # # For gluon
        # if index == 21:
    
        #     pdf = pdf / 10
    
        FFs.append(FF)

    reps.append(FFs)
```

```python
##### Plot Replicas all Together

### We have replicas, so need to figure out standard deviation ourself probably
### How to do this if have set of replicas?
### Paper says 1 sigma uncertainty


for n in range(numReps):

    plt.plot(zvalues, reps[n][iq]+reps[n][iqbar])

plt.xscale('log')
# plt.ylim((-0.01,2.0))
plt.xlim((zvalues[0], zvalues[-1]))

plt.ylabel(r'$zD^{\pi^{\pm}}_{u^{+}}(z)$', fontsize = 18)
plt.xlabel('$z $',fontsize=18)
plt.grid()
plt.title('All Replicas (NNLO)', fontsize = 20)
# plt.legend()
```

```python
### Compute Stats

zDPI_mean = 0

for n in range(numReps):

    zDPI_mean += reps[n][iq] + reps[n][iqbar]

zDPI_mean /= numReps

### Std at each z

zDPI_std = 0

for n in range(numReps):

    zDPI_std += (reps[n][iq]+reps[n][iqbar] - zDPI_mean)**2

zDPI_std = (zDPI_std / numReps)**0.5
```

```python
### Plot w/ Error Bands

plt.plot(zvalues, zDPI_mean)

plt.xscale('log')
# plt.ylim((-0.01,2.0))
plt.xlim((zvalues[0], zvalues[-1]))

plt.ylabel(r'$zD^{\pi^{\pm}}_{u^{+}}(z)$', fontsize = 18)
plt.xlabel('$z $',fontsize=18)
plt.grid()
plt.title('Mean of Replicas', fontsize = 20)
# plt.legend()
```

```python
### Plot w/ Error Bands

# Matches Figure 5.6 of https://arxiv.org/pdf/1706.07049

# plt.plot(zvalues, zDPI_mean)
plt.plot(zvalues, zDPI_mean + zDPI_std, color = 'red')
plt.plot(zvalues, zDPI_mean - zDPI_std, color = 'red')
plt.fill_between(zvalues, zDPI_mean + zDPI_std, zDPI_mean - zDPI_std, color = 'red', alpha = 0.2)

plt.xscale('log')
plt.ylim((0,12))
plt.xlim((zvalues[0], zvalues[-1]))

plt.ylabel(r'$zD^{\pi^{\pm}}_{u^{+}}(z)$', fontsize = 18)
plt.xlabel('$z $',fontsize=18)
plt.grid()
plt.title(r'$1\sigma$ Uncertainty Plot', fontsize = 20)
# plt.legend()
```

```python
### Test that mem = 0 gives Replica Average

mem0_u = reps[0][iq] + reps[0][iqbar]

plt.plot(zvalues, zDPI_mean + zDPI_std, color = 'red')
plt.plot(zvalues, zDPI_mean - zDPI_std, color = 'red')
plt.fill_between(zvalues, zDPI_mean + zDPI_std, zDPI_mean - zDPI_std, color = 'red', alpha = 0.2)

plt.plot(zvalues, mem0_u, color = 'blue')

plt.xscale('log')
plt.ylim((0,12))
plt.xlim((zvalues[0], zvalues[-1]))

plt.ylabel(r'$zD^{\pi^{\pm}}_{u^{+}}(z)$', fontsize = 18)
plt.xlabel('$z $',fontsize=18)
plt.grid()
plt.title(r'$1\sigma$ Uncertainty Plot w/ Avg', fontsize = 20)
# plt.legend()
```
